```
# Create a VideoCapture object and read from input file
# If the input is the camera, pass 0 instead of the video file name
# cap = cv2.VideoCapture('test_video.mp4')
cap = cv2.VideoCapture(VIDEO_STREAM)
```

```
person_cascade = cv2.CascadeClassifier(haarcascade_fullbody)
print(cap.get(3), cap.get(4))
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
frame_width = 480
frame_height = 320
```

```
out = cv2.VideoWriter(VIDEO_STREAM_OUT,cv2.VideoWriter_fourcc('M','J','P','G'), 20, (frame_width,frame_height))
```

```
count = 0
while(cap.isOpened()):
    r, frame = cap.read()
    if r:
        start_time = time.time()
        # frame = cv2.resize(frame,(640,360)) # Downscale to improve frame rate
        frame = cv2.resize(frame,(frame_width,frame_height)) # Downscale to improve frame rate
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY) # Haar-cascade classifier needs a grayscale image
        rects = person_cascade.detectMultiScale(gray_frame, minNeighbors=4)
        
        end_time = time.time()
        print("Elapsed Time:",end_time-start_time)
            
        for (x, y, w, h) in rects:
            cv2.rectangle(frame, (x,y), (x+w,y+h),(0,255,0),2)
        out.write(frame)

        if count >= 10000000:
            break
        
        count += 1
    
    else:
        break

    #     cv2.imshow("preview", frame)
    # k = cv2.waitKey(1)
    # if k & 0xFF == ord("q"): # Exit condition
    #     break

# When everything done, release the video capture object
cap.release()
out.release()
```
